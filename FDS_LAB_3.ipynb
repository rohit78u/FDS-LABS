```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
df =pd.read_csv("student-scores.csv")
```

```python
print("First 5 rows of dataset:")
print(df.head())
```

1. Measures of Central Tendency

```python
numeric_cols = [
    "absence_days", "weekly_self_study_hours",
    "math_score", "history_score", "physics_score",
    "chemistry_score", "biology_score",
    "english_score", "geography_score"
]
print("\n--- Measures of Central Tendency ---")
print(df[numeric_cols].mean())
print(df[numeric_cols].median())
print(df[numeric_cols].mode().iloc[0])
```

## Measures of Dispersion

```python
print("\n--- Measures of Dispersion ---")
print("Variance:\n", df[numeric_cols].var())
print("Standard Deviation:\n", df[numeric_cols].std())
print("Range:\n", df[numeric_cols].max() - df[numeric_cols].min())
```

# 3. Frequency Distribution

```python
print("\n--- Frequency Distribution (Gender) ---")
print(df['gender'].value_counts())
```

## 4. Correlation Analysis

```python
print("\n--- Correlation Matrix ---")
print(df[numeric_cols].corr())
```

## 5. Visualizations

```python
plt.subplot(2,3,1)
sns.histplot(df['math_score'], bins=20, kde=True)
plt.title("Histogram of Math Score")

# Boxplot
plt.subplot(2,3,2)
sns.boxplot(y=df['math_score'])
plt.title("Boxplot of Math Score")

```

```python
# Bar Chart
plt.subplot(2,3,3)
df['gender'].value_counts().plot(kind='bar', color=['skyblue','pink'])
plt.title("Bar Chart of Gender")
```

```python

# Pairplot
sns.pairplot(df[numeric_cols[:5]], diag_kind="kde")
plt.suptitle("Pairplot of Selected Features", y=1.02)
```

```python
# Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Heatmap of Correlation")
plt.show()
```

## 6. Skewness & Kurtosis

```python
print("\n--- Skewness & Kurtosis ---")
print("Skewness:\n", df[numeric_cols].skew())
print("Kurtosis:\n", df[numeric_cols].kurt())
```

## 7. Percentiles & Quartiles

```python
print("\n--- Percentiles & Quartiles (Math Score) ---")
print("25th Percentile (Q1):", np.percentile(df['math_score'], 25))
print("50th Percentile (Median):", np.percentile(df['math_score'], 50))
print("75th Percentile (Q3):", np.percentile(df['math_score'], 75))
```

## 8. Coefficient of Variation

```python
cv = df[numeric_cols].std() / df[numeric_cols].mean()
print("\nCoefficient of Variation:\n", cv)
```

## 9. Z-scores


```python
df_zscores = df[numeric_cols].apply(stats.zscore)
print("\nFirst 5 rows of Z-scores:\n", df_zscores.head())
```

## 10. Interquartile Range (IQR)

```python
IQR = df[numeric_cols].quantile(0.75) - df[numeric_cols].quantile(0.25)
print("\nInterquartile Range (IQR):\n", IQR)
```

## 11. Cross-tabulation

```python
print("\n--- Cross Tabulation (Gender vs Career Aspiration) ---")
print(pd.crosstab(df['gender'], df['career_aspiration']))
```
